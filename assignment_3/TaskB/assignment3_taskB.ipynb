```python
import matplotlib.pyplot as plt
%matplotlib widget
import numpy as np
import sys
import taurex.log
taurex.log.disableLogging()

from taurex.model import TransmissionModel
from taurex.planet import Planet
from taurex.stellar import BlackbodyStar
from taurex.chemistry import TaurexChemistry, ConstantGas
from taurex.temperature import Guillot2010
```

```python
data= np.genfromtxt('table_WASP-107-b-Welbanks-et-al.-2024.csv',delimiter=',', dtype='float',usecols=(0,2,3),unpack=True, skip_header=1)
np.shape(data)

wavelength = data[0,:]
opt_depth = 10**-2*data[1,:]
err = 10**-2*data[2,:]

plt.figure()
plt.errorbar(wavelength,opt_depth,yerr=err)
plt.title('Absorption Profile of the Planet WASP-107 b')
plt.xlabel('Wavelength ($\mu$m)')
plt.ylabel('Absorbed relative flux')
#plt.xscale('log')
plt.grid(linestyle='--')
plt.savefig('WASP_107b_spectrum_assignment3.png', dpi=300)

plt.show()

```

```python
planet_name = "WASP-107-b"
output_file = f"{planet_name}_spectrum_assignment3_taskB.dat"

with open(output_file, 'w') as f:
    f.write("# Wavelength (micron), (rp/rs), Error\n")
    for wl, rprs2, error in zip(wavelength, opt_depth, err):
        f.write(f"{wl:.6f} {rprs2:.6e} {error:.6e}\n")

print(f"Transmission spectrum saved to {output_file}")
```
